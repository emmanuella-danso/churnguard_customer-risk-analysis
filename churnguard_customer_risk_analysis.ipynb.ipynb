```python
import numpy as np
import pandas as pd

rows = []

for _, row in df.iterrows():
    customer_id = row['customerID']
    tenure = int(row['tenure'])
    base_charge = row['MonthlyCharges']
    
    # Assign behavior
    behavior = np.random.choice(
        ['stable', 'decline', 'volatile', 'pre_churn'],
        p=[0.4, 0.3, 0.2, 0.1]
    )
    
    for month in range(1, tenure + 1):
        
        if behavior == 'stable':
            spend = base_charge * np.random.uniform(0.95, 1.05)
        
        elif behavior == 'decline':
            decay = 1 - (month / (tenure * 1.5))
            spend = base_charge * decay * np.random.uniform(0.95, 1.05)
        
        elif behavior == 'volatile':
            spend = base_charge * np.random.uniform(0.7, 1.3)
        
        elif behavior == 'pre_churn':
            if month > tenure * 0.7:
                spend = base_charge * np.random.uniform(0.4, 0.7)
            else:
                spend = base_charge * np.random.uniform(0.95, 1.05)
        
        rows.append([
            customer_id,
            month,
            round(spend, 2),
            behavior
        ])

time_df = pd.DataFrame(rows, columns=[
    'customerID',
    'month',
    'monthly_spend',
    'behavior_type'
])

time_df.head()
```

```python
time_df['rolling_avg_spend'] = (
    time_df.groupby('customerID')['monthly_spend']
    .transform(lambda x: x.rolling(window=3, min_periods=1).mean())
)
```

```python
time_df['spend_change'] = (
    time_df.groupby('customerID')['rolling_avg_spend']
    .diff()
)
```

```python
time_df['decline_flag'] = time_df['spend_change'] < 0
```

```python
time_df['revenue_at_risk'] = np.where(
    time_df['decline_flag'],
    abs(time_df['spend_change']),
    0
)
```

```python
customer_risk = time_df.groupby('customerID').agg({
    'revenue_at_risk': 'sum'
}).reset_index()
```

```python
time_df.head(10)
```

```python
customer_summary = time_df.groupby('customerID').agg({
    'monthly_spend': 'mean',
    'revenue_at_risk': 'sum',
    'decline_flag': 'sum'
}).reset_index()
```

```python
customer_summary['risk_score'] = (
    customer_summary['revenue_at_risk'] * 0.6 +
    customer_summary['monthly_spend'] * 0.3 +
    customer_summary['decline_flag'] * 0.1
)
```

```python
customer_summary = customer_summary.sort_values(
    by='risk_score',
    ascending=False
)
```

```python
customer_summary.head(10)
```

```python
customer_summary['action_segment'] = pd.cut(
    customer_summary['risk_score'],
    bins=3,
    labels=['Monitor', 'At Risk', 'Urgent Action']
)
```

```python
customer_summary['action_segment'].value_counts()
```

```python
def assign_action(row):
    if row['action_segment'] == 'Urgent Action':
        return 'Call / Retain Offer / Discount Intervention'
    elif row['action_segment'] == 'At Risk':
        return 'Engagement Campaign / Usage Incentive'
    else:
        return 'Monitor - No Action'

customer_summary['recommended_action'] = customer_summary.apply(assign_action, axis=1)
```
